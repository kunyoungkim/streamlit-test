```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
print(tips.head())
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))
sns.barplot(x="day", y="total_bill", data=tips, ax = ax[0, 0])
ax[0, 0].set_title("title 0, 0")
sns.scatterplot(x="total_bill", y="tip", data=tips, ax = ax[0, 1])
sns.pointplot(x="day", y="tip", data=tips, ax = ax[1, 0])
sns.stripplot(x="day", y="total_bill", data=tips, ax = ax[1, 1])

plt.tight_layout()
plt.show()
```
